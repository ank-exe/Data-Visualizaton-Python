```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from GitHub
data_url = 'https://raw.githubusercontent.com/premsai1307/game-report-/refs/heads/main/Video%20Games%20Sales.csv'  # Replace with the actual raw GitHub URL
data = pd.read_csv(data_url)

# Handle missing values in Year
data['Year'] = data['Year'].fillna(0).astype(int)

# 1. Top 10 Games by Global Sales
top_games = data.nlargest(10, 'Global')
plt.figure(figsize=(10, 6))
plt.bar(top_games['Game Title'], top_games['Global'], color='gold')
plt.title('Top 10 Games by Global Sales', fontsize=16)
plt.xlabel('Game Title', fontsize=12)
plt.ylabel('Global Sales (in millions)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# 2. Sales Distribution Across Regions
regions = ['North America', 'Europe', 'Japan', 'Rest of World']
region_sales = data[regions].sum()
plt.figure(figsize=(10, 6))
plt.bar(region_sales.index, region_sales.values, color=['blue', 'orange', 'green', 'red'])
plt.title('Sales Distribution Across Regions', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Sales (in millions)', fontsize=12)
plt.tight_layout()
plt.show()

# 3. Year-on-Year Growth of Global Sales
year_sales = data[data['Year'] > 0].groupby('Year')['Global'].sum()
plt.figure(figsize=(10, 6))
plt.plot(year_sales.index, year_sales.values, marker='o', color='purple')
plt.title('Year-on-Year Growth of Global Sales', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Global Sales (in millions)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

# 4. Genre Popularity by Region
genre_region = data.groupby('Genre')[regions].sum()
ax = genre_region.plot(kind='bar', stacked=False, figsize=(12, 6), colormap='tab10')
plt.title('Genre Popularity by Region', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Sales (in millions)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Region', fontsize=10)
plt.tight_layout()
plt.show()

# 5. Publisher Contribution to Global Sales
top_publishers = data.groupby('Publisher')['Global'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
plt.pie(top_publishers, labels=top_publishers.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 10 Publishers Contribution to Global Sales', fontsize=16)
plt.tight_layout()
plt.show()

# 6. Regional Sales Correlation
plt.figure(figsize=(8, 6))
plt.scatter(data['North America'], data['Europe'], color='teal', alpha=0.5)
plt.title('Correlation: North America vs. Europe Sales', fontsize=16)
plt.xlabel('North America Sales (in millions)', fontsize=12)
plt.ylabel('Europe Sales (in millions)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

# 7. Platform Lifecycle Analysis
platform_year = data.groupby(['Platform', 'Year'])['Global'].sum().unstack(fill_value=0)
platform_year.T.plot(kind='area', stacked=True, figsize=(12, 6), colormap='Spectral')
plt.title('Platform Lifecycle Analysis', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Global Sales (in millions)', fontsize=12)
plt.legend(title='Platform', fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```
